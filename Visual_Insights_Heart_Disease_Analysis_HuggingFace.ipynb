```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = "https://huggingface.co/datasets/mstz/heart_failure/resolve/main/heart_failure_clinical_records_dataset.csv"
df = pd.read_csv(url)
```

```python
print("First 5 rows of the dataset:\n", df.head())
print("\nDataset Info:")
df.info()
print("\nMissing Values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution of Heart Failure Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=df, palette='Set2')
plt.title("Serum Creatinine Levels by Death Event")
plt.xlabel("Death Event (0 = Survived, 1 = Died)")
plt.ylabel("Serum Creatinine (mg/dL)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig = px.scatter(df,
                 x='age',
                 y='ejection_fraction',
                 color='DEATH_EVENT',
                 title="Age vs. Ejection Fraction Colored by Death Event",
                 labels={'DEATH_EVENT': 'Death Event'},
                 hover_data=['serum_creatinine', 'platelets'])
fig.show()
```

```python
print("\n--- Inference ---")
print("1. Most patients fall in the 50-70 age range.")
print("2. Higher serum creatinine levels are observed among deceased patients.")
print("3. Patients with lower ejection fraction are more likely to have experienced death events.")
```
